# Assignment 2 â€” Visualizing a Neural Network Architecture (Fraud Detection)


```python
import matplotlib.pyplot as plt

def draw_fraud_neural_network():
    # Define layers [Input, Hidden, Output]
    # 6 Features -> 4 Hidden Neurons -> 1 Fraud Probability
    layers = [6, 4, 1]
    feature_names = [
        "Transaction Amount ", "Transaction Time", "Merchant Category",
        "Customer Age", "Account Balance", "Transactions Today"
    ]

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.axis('off')

    # Position calculations
    v_spacing = 1.0 / max(layers)
    h_spacing = 1.0 / len(layers)

    # Draw Nodes
    for i, layer_size in enumerate(layers):
        layer_top = v_spacing * (layer_size - 1) / 2.0
        for j in range(layer_size):
            x = i * h_spacing + 0.1
            y = layer_top - j * v_spacing + 0.5

            circle = plt.Circle((x, y), 0.03, color='#3498db', ec='black', zorder=4)
            ax.add_artist(circle)

            # Label Input Features
            if i == 0:
                ax.text(x-0.05, y, feature_names[j], ha='right', va='center', fontsize=9)
            # Label Output
            if i == len(layers) - 1:
                ax.text(x+0.05, y, "Fraud Probability", ha='left', va='center', fontweight='bold')

    # Draw Edges (Connections)
    for i in range(len(layers) - 1):
        layer_top_curr = v_spacing * (layers[i] - 1) / 2.0
        layer_top_next = v_spacing * (layers[i+1] - 1) / 2.0
        for j in range(layers[i]):
            for k in range(layers[i+1]):
                line = plt.Line2D(
                    [i * h_spacing + 0.1, (i + 1) * h_spacing + 0.1],
                    [layer_top_curr - j * v_spacing + 0.5, layer_top_next - k * v_spacing + 0.5],
                    color='gray', lw=0.5, alpha=0.5, zorder=1
                )
                ax.add_artist(line)

    plt.title("Credit Card Fraud Detection Architecture", pad=20)
    plt.show()
```

```python
draw_fraud_neural_network()
```
